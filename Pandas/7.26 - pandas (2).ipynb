```python
#시리즈와 데이터프레임이 지원하는 기능은 다르다, 시리즈와 비슷해보이지만, 데이터프레임으로 가져오기위해서는 리스트형식으로 넣어야한다.
#관례상 어트리뷰트 방식보다 , 딕셔너리 방식이 조금더 좋다
#어트리뷰트 방식은 df.컬럼명 이기 때문에 , 하나밖에못가져옴, 
#이에 비해 딕셔너리방식은 순서도 지원하고, 여러개 가져올수있음
```

```python
df[0] # 딕셔너리는 컬럼명으로 뽑는 방식인데 컬렴명에 0 이없으니 당연히 에러.
```

```python
df.loc[0]
```

```python
# loc 는 row. loc 없이는 column. 
```

```python
df.loc[[0,2,4]] #pandas 에서는 리스트형식으로 넣어야 동시에 여러값을 가져옴
```

```python
#row 는 loc 쓴다.
```

```python
df.loc[0:4]
```

```python
#컬럼은 slicing  이 안된다 . 딕셔너리 구조라서 , 
#하지만 row 는 slicing  이 된다 . 

df.loc[1:3,'title']  #2차원 slicing title 열 1에서 3 까지 가져온다
```

```python
df.loc[1:3,['title','rating']] #두개이상 가져올때는 리스트를 쓴다 .
```

```python
df.loc[:,["title","rating"]] # 이건 딕셔너리방식하고 같다 
```

```python
# 위의것은 데이터프레임이니까, 함수 적용가능
```

```python
# loc 가 있고  iloc 가있다. i = index 베이스, l = label 인덱스
```

```python
%timeit df.at[1,'rating'] #이거는 하나만 가져올때 ,인덱싱하는것, 슬라이싱 X 
```

```python
%timeit df.loc[1,'rating'] # 이거는 범위를 가져올 수 있음. 
```

```python
#성능 체크 , at은 속도가 빠르다 , loc 이 조금 느리다
```

```python
df.loc[1,'rating']  = 19 # 이렇게 값을 바꿀 수 있다, 수정할때 사용가능 ,
```

```python
df.axes
```

```python
df.iat[1,1] # 앞의 것은 row , 뒤의것은 column 으로본다. i  = index based
```

```python
df.at[1,1] #  이것은 label 베이스 이기 때문에 row 와 columng 에 대한 index 를 넣으면 이렇게 에러가 발생한다 .
```

```python
timeit df.iat[1,1]
```

iat 은 변환단계를 거쳐서 조금느림 하지만 눈에 보이기 좋아서 iat 을 훨씬더 많이씀

```python
%timeit df.at[1,'title']
```

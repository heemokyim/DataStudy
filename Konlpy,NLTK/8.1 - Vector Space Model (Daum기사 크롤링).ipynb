```python
uniqNouns  =  []# 모든 문서에 나온 유니크한 명사 이것이 문서를 표현하는 벡터가 될 것이다 .


for file in fileList:
    docid=file[-17:]
    
    with open(file,"r",encoding='utf-8') as fp:
        content = fp.read()
        content = re.sub(r"[\s]{2,}","",content)  # white space 가 최소 2개이상 붙은것들을 없에준다.

    for sentence in nltk.sent_tokenize(content):
        nouns = []

        for word in nltk.word_tokenize(sentence):# 어절단위로 분리된것을 가지고 noun 을 만들것이다.
            nouns.extend(Kkma().nouns(word))
            #uniq 만 뽑아야하니까 초기화해준다. nouns 에는 kkma 를 써서 단어로 자른게 들어가있다.
            nouns = list(set(nouns)) # set 해서 unique 하게   + 반환이 셋형이라 list 로 묶는다.
        uniqNouns.extend(nouns)
        uniqNouns = list (set(uniqNouns))
        #print(sentence)
        #print(nouns)

print('uniq' ,len(uniqNouns))
```

```python
docNouns  =  {}# 모든 문서에 나온 유니크한 명사 이것이 문서를 표현하는 벡터가 될 것이다 .
maxCount= {}

for file in fileList:
    docid=file[-17:]
    
    with open(file,"r",encoding='utf-8') as fp:
        maxCount[docid]=0 #초기화
        content = fp.read()
        content = re.sub(r"[\s]{2,}","",content)  # white space 가 최소 2개이상 붙은것들을 없에준다.
    nouns = {}
    for sentence in nltk.sent_tokenize(content):
        for word in nltk.word_tokenize(sentence):# 어절단위로 분리된것을 가지고 noun 을 만들것이다.
            for w in Kkma().nouns(word):
                if w in nouns.keys():
                    nouns[w]+=1
                else:
                    nouns[w] = 1
                if maxCount[docid] < nouns[w]: #비교를 위해서 밑에서 초기화.
                    maxCount[docid] =nouns[w]
    docNouns[docid] = nouns
```

# 어떤 단어가 어떤 문서에서 나왔는지,찾아주기. 인덱스를 만들기( 책 맨뒤)

```python
# 인덱싱 작업
invertedIndex = {}

for noun in uniqNouns:
    invertedIndex[noun] =  [] # 초기화.
    
    for k,v in docNouns.items():# 키값이 docid 값이 명사랑 벨류값
        if noun in v:
            invertedIndex[noun].append(k) #해당 단어 key 값에 대해 단어가 속한 문서를 value 값으로준다.
```

docNouns 에 TF 값이 있다.

```python
kRation  =0.5
TF = {}
for k , v in docNouns.items():
    tfList = {}
    for w in v:
        tfList[w] = kRation + ( 1- kRation ) * (v[w] / maxCount[k]) 
        print("{0} | {1} + {2} *({3}/{4}) ={5}".format(w,kRation,(1-kRation),v[w],maxCount[k],tfList[w]))
        #v[w] # 특정 문서에 있는 단어만 나옴.
        # maxCOunt 에는 키 가 특정문서.
    TF[k] = tfList # k 는 문서의 id , 내가만든 리스트를 저장.
```

```python
from math import log10 # log 쓰기위해서
```

```python
docSize = len(fileList)
TFIDF = {}

for k,v in TF.items(): # k 가 docid, 
    idfList = {}
    for t in v: # t가 keyword 다. 
        idf = log10(docSize/len(invertedIndex[t]))
        idfList[t] = v[t] * idf # TF * IDF 값
        print("{0}|{1}*log({2}/{3})={4}".format(t,v[t],docSize,len(invertedIndex[t]),idfList[t]))
    TFIDF[k] = idfList
```

```python
query = "소비자에게 신차값 수준의 보상금 BMW 화재"

queryNouns = {}
maxCount = 0

for word in nltk.word_tokenize(query): # 한 문장을 whitespace, 특수문자 기준으로 tokenize된 단어 for문
    for w in Kkma().nouns(word): # 명사들에 대해서 for문
        if w in queryNouns.keys(): 
            queryNouns[w] += 1
        else:
            queryNouns[w] = 1
        if maxCount < queryNouns[w]: # 명사의 빈도수가 현재 maxcount에 저장되어 있는 값보다 크다면
            maxCount = queryNouns[w] # 명사 빈도수의 큰 값을 갱신해줌
    
print(len(queryNouns), maxCount)
print(queryNouns)
```

```python
queryWeight = {}
for k,v in queryNouns.items():
    queryWeight[k] = kRation + (1-kRation) * (v/maxCount)
    print("{0} | {1} + {2} *({3}/{4}) ={5}".format(k,kRation,(1-kRation),v,maxCount,queryWeight[k]))
```

# 쿼리도 새로운 형태의 벡터가 됨.

벡터의 크기를 구해야한다. 다큐먼트에 포함되어있는 모든 색인어들(명사)가중치의 제곱의합에 루트 . 

```python
from math import sqrt
docLength = {} # 문서 1 : 길이, 문서2  : 길이...

for k,v in TFIDF.items():
    sumPow = 0.0
    
    for t in v:
        sumPow += v[t]**2 
    
    docLength[k] = sqrt(sumPow)
    
 #   k #문서ID
  #  v #단어1:가중치,단어2:가중치...
```

```python
print(docLength)
```

```python
result = {} 
for k,v in  queryWeight.items():
    if k in invertedIndex.keys():
      #  print(k)
        for docid in invertedIndex[k]:
            if docid in result.keys():
                result[docid] +=TFIDF[docid][k] * queryWeight[k]
            else:
                result[docid] =TFIDF[docid][k] * queryWeight[k]
                
        #   print(" - DOCID : {0},Weight:{1}".format(docid,TFIDF[docid][k]))
            result[docid] = TFIDF[docid][k] * queryWeight[k]
            
print(result) # 각문서들마다 쿼리랑 비교했던 값들의 summation

for k,v in result.items():
    result[k] = v / docLength[k]


result = dict(sorted(result.items(), key = lambda k:k[1],reverse =True))

for k,v in result.items():
    print("[{0}] 유사도 {1}".format(k,v))
    with open("C:\\Users\\user\\dev\\bigdata\\8.1 nltk\\data\\" + k , "r" , encoding='utf8') as fp:
        print(fp.readlines(10))
```

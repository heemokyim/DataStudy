```python
uniqNouns  =  []# 모든 문서에 나온 유니크한 명사 이것이 문서를 표현하는 벡터가 될 것이다 .


for file in fileList:
    docid=file[-17:]
    
    with open(file,"r",encoding='utf-8') as fp:
        content = fp.read()
        content = re.sub(r"[\s]{2,}","",content)  # white space 가 최소 2개이상 붙은것들을 없에준다.

    for sentence in nltk.sent_tokenize(content):
        nouns = []

        for word in nltk.word_tokenize(sentence):# 어절단위로 분리된것을 가지고 noun 을 만들것이다.
            nouns.extend(Kkma().nouns(word))
            #uniq 만 뽑아야하니까 초기화해준다.
            nouns = list(set(nouns)) # set 해서 unique 하게   + 반환이 셋형이라 list 로 묶는다.
        uniqNouns.extend(nouns)
        uniqNouns = list (set(uniqNouns))
#         print(sentence)
    #         print(nouns)

print(len(uniqNouns))
```

```python
docNouns  =  {}# 모든 문서에 나온 유니크한 명사 이것이 문서를 표현하는 벡터가 될 것이다 .
maxCount= {}

for file in fileList:
    docid=file[-17:]
    
    with open(file,"r",encoding='utf-8') as fp:
        maxCount[docid]=0 #초기화
        content = fp.read()
        content = re.sub(r"[\s]{2,}","",content)  # white space 가 최소 2개이상 붙은것들을 없에준다.
    nouns = {}
    for sentence in nltk.sent_tokenize(content):
        for word in nltk.word_tokenize(sentence):# 어절단위로 분리된것을 가지고 noun 을 만들것이다.
            for w in Kkma().nouns(word):
                if w in nouns.keys():
                    nouns[w]+=1
                else:
                    nouns[w] = 1
                if maxCount[docid] < nouns[w]: #비교를 위해서 밑에서 초기화.
                    maxCount[docid] =nouns[w]
    docNouns[docid] = nouns
```

```python
for k,v in docNouns.items():
    print("{0} - {1}/{2}".format(k,len(v),maxCount[k])) #문서명, 각 문서에 대해서 단어의 수, 단어중 최대 빈도수
```

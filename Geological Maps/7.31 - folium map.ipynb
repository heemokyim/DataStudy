```python
m9  = folium.Map(location=[37.5665,126.9780],tiles='Mapbox Bright')
```

```python
seoul_geo_json=r'seoul_municipalities_geo_simple.json'
```

```python
seoul_topo_json=r'seoul_municipalities_topo_simple.json' #r은 관례상 raw 파일 나타내는 표시
```

```python
folium.TopoJson(open(seoul_topo_json,encoding='utf-8'),"objects.seoul_municipalities_geo",name="topojson",).add_to(m9)
```

```python
folium.GeoJson(open(seoul_geo_json,encoding='utf-8').read(),"objects.seoul_municipalities_geo",name="geojson").add_to(m9)
```

```python
m99 = folium.Map(location = [37.5665, 126.9780], tiles = 'Mapbox Bright')

seoul_geo_json = r'seoul_municipalities_geo_simple.json'
seoul_topo_json = r'seoul_municipalities_topo_simple.json'

folium.GeoJson(open(seoul_geo_json, encoding='utf-8-sig').read(), name='geojson').add_to(m99)
folium.TopoJson(open(seoul_topo_json, encoding='utf-8-sig'), "objects.seoul_municipalities_geo", name='topojson', ).add_to(m99)

folium.LayerControl().add_to(m99)
```

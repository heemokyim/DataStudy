```python
from matplotlib import pyplot as plt
```

```python
from konlpy.corpus import kolaw
```

```python
from konlpy.corpus import kobill
```

```python
kobill.fileids()
```

```python
text = [kobill.open(i).read() for i in kobill.fileids()]
```

```python
# 10개 짜리 텍스트 리스트가 생긴다.
```

```python
len(text)
```

```python
global_unique = []
global_unique_count = []
```

```python
for row in text:
    term = t.nouns(row)
    unique = set(term)
    global_unique  += list(unique)
    global_unique =  list(set(global_unique))
    global_unique_count.append(len(global_unique))
    print(len(unique),len(global_unique))
```

```python
plt.plot(global_unique_count)
plt.savefig("test.png")
```
